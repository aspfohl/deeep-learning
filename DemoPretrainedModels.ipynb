```python
import torch
from dalle_pytorch import OpenAIDiscreteVAE, DALLE
import datetime

# from the main readme directly
# https://github.com/lucidrains/DALLE-pytorch

vae = OpenAIDiscreteVAE()       # loads pretrained OpenAI VAE

start = datetime.datetime.now()

dalle = DALLE(
    dim = 1024,
    vae = vae,                  # automatically infer (1) image sequence length and (2) number of image tokens
    num_text_tokens = 10000,    # vocab size for text
    text_seq_len = 256,         # text sequence length
    depth = 1,                  # should aim to be 64
    heads = 2,                  # attention heads (changed from 16)
    dim_head = 64,              # attention head dimension
    attn_dropout = 0.1,         # attention dropout
    ff_dropout = 0.1            # feedforward dropout
)

text = torch.randint(0, 10000, (4, 256))

images = dalle.generate_images(
    text,
    cond_scale = 3. # secondly, set this to a value greater than 1 to increase the conditioning beyond average
)
print(images.shape)
print((datetime.datetime.now() - start).seconds)
# took 2 HOURS!
```

```python
import torch
from matplotlib import pyplot as plt 
from dalle_pytorch import VQGanVAE, DALLE
import datetime

# Also from the main dalle readme, this uses a different variational auto encoder
# https://github.com/lucidrains/DALLE-pytorch
start = datetime.datetime.now()
print((datetime.datetime.now() - start).seconds)
vae = VQGanVAE()

dalle = DALLE(
    dim = 1024,
    vae = vae,                  # automatically infer (1) image sequence length and (2) number of image tokens
    num_text_tokens = 10000,    # vocab size for text
    text_seq_len = 256,         # text sequence length
    depth = 1,                  # should aim to be 64
    heads = 16,                 # attention heads
    dim_head = 64,              # attention head dimension
    attn_dropout = 0.1,         # attention dropout
    ff_dropout = 0.1            # feedforward dropout
)

text = torch.randint(0, 10000, (4, 256))

images2 = dalle.generate_images(
    text,
    cond_scale = 3. # secondly, set this to a value greater than 1 to increase the conditioning beyond average
)
print((datetime.datetime.now() - start).seconds)
images2.shape
```
